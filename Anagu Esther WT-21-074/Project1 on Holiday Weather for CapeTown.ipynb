# Cleaning the data

*....Before using the dataset, it is important to clean the data and fix irregilarities, such as missing data, incomplete data, and inaccurate data in order to produce a credible and accurate analysis......*

To do this, we will remove the < br /> html line breaks from the values in the 'WindDirDegrees' column and change the values to flats.

```python
CapeTown = CapeTown.rename(columns={'WindDirDegrees<br />' : 'WindDirDegrees'})
CapeTown
```

As you can see, the WinDirDegrees column does not have < br /> html line breaks anymore. As individually done below:

```python
CapeTown['WindDirDegrees'] = CapeTown['WindDirDegrees'].str.rstrip('<br />')
CapeTown['WindDirDegrees']
```

```python
CapeTown['WindDirDegrees'] = CapeTown['WindDirDegrees'].astype('float64')
CapeTown['WindDirDegrees']
```

We definitely need to change the values in the 'Date' column into values of the datetime64 date type. We also need to change the index from the default to the `datetime64` values in the `'Date'` column so that it is easier to pull out rows between particular dates and display more meaningful graphs: 

```python
CapeTown['Date'] = pd.to_datetime(CapeTown['Date'])
CapeTown['Date']
```

```python
CapeTown.index = CapeTown['Date']
CapeTown.index
```

## Finding a summer break

According to Alexander Hamilton(2021), "summer in CapeTown usually starts from mid-October all through mid-February." However, we are going to create a dataframe that holds just three months: October, November, and December, using the datetime index, like the one below:

```python
summer = CapeTown.loc['2014-10-14' : '2014-12-31']
summer
```

It is time for us to look for the days with warm temperatures.

```python
summer[summer['Mean TemperatureC'] >= 25]
```

Summer 2014 was cool in CapeTown: there are no days with temperatures of 25 Celsius or higher except the 31st of December, 2014. Best to see a graph of the temperature and look for the warmest period.

So the next thing we will do is to display a graph to visualize the above result.

We will plot the 'Mean TemperatureC' for the summer:

```python
%matplotlib inline
```

```python
summer['Mean TemperatureC'].plot(grid=True, figsize=(10,5))
```

Looking at the above graph, the end of December looks good for mean temperatures over 23 degrees C. We will also put precipitation on the graph too:

```python
summer[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

Let's have a closer look by just plotting mean temperature and precipitation for December.

```python
December = summer.loc['2014-12-1' : '2014-12-31']
December[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

# Conclusion

Yes, the month of December looks very good. However, there was rain, though not significant, in just three days: 12th, 21st and the 25th. Besides, there was no day that the mean temperature dropped below 19 degrees. Although, there were days with 19 degrees: 4th, 12th, 16th, 21th, and 26th.

The graphs have shown the volatility of a CapeTown summer. A couple of weeks were found when the weather wasn't too bad in 2014. Of course this is no guarantee that the weather pattern will repeat itself in future years. To make a sensible prediction we would need to analyse the summers for many more years.
