# Study of GDP and Life Expectancy

Two datasets of the World Bank are considered. One dataset, available at http://data.worldbank.org/indicator/NY.GDP.MKTP.CD, lists the GDP of the world's countries in current US dollars, for various years. The other dataset, available at http://data.worldbank.org/indicator/SP.DYN.LE00.IN, lists the life expectancy of the world's countries. The datasets were downloaded as CSV files in June 2021 and we considered the year 2017 in this analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
all_data_life=pd.read_csv('API_SP.DYN.LE00.IN_DS2_en_csv_v2_2445326.csv')
all_data_life.head()
```

```python
life_expectancy=pd.DataFrame(all_data_life,columns=['Country Name','2017',])
life_expectancy.head()
```

```python
life_expectancy=life_expectancy.rename(columns={'2017':'life_expectancy(years)'})
life_expectancy.head()
```

```python
all_data_gdp=pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_2445719.csv')
all_data_gdp.head()
```

```python
gdp_data=pd.DataFrame(all_data_gdp,columns=['Country Name','2017'])
gdp_data.head()
```

```python
gdp_data=gdp_data.rename(columns={'2017':'gdp_indicator'})
gdp_data.head()
```

## Data Transformation

The price of dollar to naira at 17:35 GMT of 17/06/2021 is 410 naira to a dollar which we would convert to for the purpose of this study.

```python
def round_to_millions(value):
    return round(value/100000)
def usd_to_naira(usd):
    return usd*410
```

```python
gdp_data["GDP (naira)"]=gdp_data['gdp_indicator'].apply(round_to_millions).apply(usd_to_naira)
gdp_data.head()
```

```python
gdp_data.drop("gdp_indicator",inplace=True,axis=1)
```

```python
gdp_data
```

```python
life_expectancy.head()
```

```python
life_expectancy.shape,gdp_data.shape
```

There are more countries in the gdp table than in the life expectancy table so we perform an inner join.

```python
gdp_vs_life=pd.merge(gdp_data,life_expectancy,on="Country Name",how='inner')
print(gdp_vs_life.shape)
gdp_vs_life.head()
```

## Calculating the Correlation

```python
from scipy.stats import spearmanr
```

```python
gdp_column=gdp_vs_life['GDP (naira)']
life_exp_column=gdp_vs_life['life_expectancy(years)']
```

```python
(correlation,pValue)=spearmanr(gdp_column,life_exp_column)
print(f'The correlation is ', correlation)
```

```python
if pValue < 0.05:
    print('It is statistically significant.')
else:
    print('It is not statistically significant.')
```

Our value, 0.3 is closer to 1 than it is to zero. So, this means that the two indices are directly proportional to one another. That is, richer countries tend to have longer life expectancy.
But we see here that the correlation is a weak correlation.

```python
gdp_vs_life.plot(x="GDP (naira)", y='life_expectancy(years)',
                 kind='scatter',logx=True, figsize=(10, 6),grid=True)
plt.title("GDP versus Life Expectancy",fontweight="bold",fontsize=20)
plt.savefig("GDP versus Life Expectancy.png")
plt.show()
```

```python
gdp_vs_life.corr()
```

Both the plot and this table shows that there is very weak correlation between life expecatancy and GDP in the year 2017.<br>

From the graph, we see that some countries with low GDP have high life expectancy and vice-versa. So, we can say that other factors are responsible for high life expectancy.

## Conclusion

The spearman's coefficent shows that there is weak correlation between life expectancy and GDP. However, a careful look at the countries with the least life expectancy shows that the top countries with the least life expectancy are African countries.<br>

Furthermore, few of these countries still find themselves among the top 30 countries with low GDP.<br>

Hence, we conclude that even if there is no strong correlation between life expectancy and GDP, countries in Africa have less life expectancy and then we can further research about other factors that may be responsible for life expectancy.
