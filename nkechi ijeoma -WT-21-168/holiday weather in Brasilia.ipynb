```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

import pandas as pd
Brasilia = pd.read_csv('Brasilia_BSB_2014.csv', skipinitialspace=True)
from datetime import datetime
```

```python
#clean data

Brasilia.head
```

```python
Brasilia= Brasilia.rename(columns={'WindDirDegrees<br />' : 'WindDirDegrees'})
```

```python
Brasilia['WindDirDegrees'] = Brasilia['WindDirDegrees'].str.rstrip('<br />')
```

```python
Brasilia.dtypes
```

```python
Brasilia['WindDirDegrees'] = Brasilia['WindDirDegrees'].astype('float64')   
```

```python
Brasilia.dtypes
```

```python
Brasilia['Date'] = pd.to_datetime(Brasilia['Date'])
```

```python
Brasilia.index = Brasilia['Date']
```

```python
#summer break
summer = Brasilia.loc[datetime(2014,1,1) : datetime(2014,3,16)]
```

```python
summer[summer['Mean TemperatureC'] >= 25]
```
