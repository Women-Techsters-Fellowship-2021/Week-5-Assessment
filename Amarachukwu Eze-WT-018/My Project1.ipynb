```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

import pandas as pd
london = pd.read_csv('London_2014.csv', skipinitialspace=True) #load the CSV file into a dataframe
```

```python
london = london.rename(columns={'WindDirDegrees' : 'WindDirDegrees'})
```

```python
london['WindDirDegrees'] = london['WindDirDegrees'].str.rstrip('<br />')#remove the < br /> html line breaks from the values in the 'WindDirDegrees' column.
```

```python
london['WindDirDegrees'] = london['WindDirDegrees'].astype('float64') #change the values in the 'WindDirDegrees' column to float64
```

```python
london['GMT'] = pd.to_datetime(london['GMT'])#change the values in the 'GMT' column into values of the datetime64 date type.
```

```python
london.index = london['GMT']#change the index from the default to the datetime64 values in the 'GMT' column so that it is 
#easier to pull out rows between particular dates and display more meaningful graphs:
```

```python
summer = london.loc[pd.datetime(2014,6,1) : pd.datetime(2014,8,31)]#create a dataframe that holds just the 3 months (june,july &august) using the datetime index
```

```python
summer[summer['Mean TemperatureC'] >= 25]#look for the days with warm temperatures.
```

```python
%matplotlib inline #to display any graph created inside this notebook
```

```python
summer['Mean TemperatureC'].plot(grid=True, figsize=(10,5)) #plots the 'Mean TemperatureC' for the summer
```

```python
summer[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))#puts precipitation on the graph
```

```python
#mean temperature and precipitation for July.
july = summer.loc[pd.datetime(2014,7,1) : pd.datetime(2014,7,31)]
july[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```
