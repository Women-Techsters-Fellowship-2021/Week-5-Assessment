```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

import pandas as pd
from datetime import datetime 
moscow = pd.read_csv('Moscow_SVO_2014.csv')
```

```python
moscow.head()
```

```python
moscow.tail()
```

```python
#renaming the wind degrees columns
moscow = moscow.rename(columns={'WindDirDegrees<br />' : 'WindDirDegrees'})
```

```python
moscow.head()
```

```python
#removing the html tags
moscow['WindDirDegrees'] = moscow['WindDirDegrees'].str.rstrip('<br />')
```

```python
moscow.head()
```

```python
#changing the values to float from string
moscow['WindDirDegrees'] = moscow['WindDirDegrees'].astype('float64')  
```

```python
#changing date to datetime
moscow['Date'] =pd.to_datetime(moscow['Date'])
```

```python
#changing the index column to Date
moscow.index = moscow['Date']
```

```python
#choosing the 12 months  period in moscow to just see the overall weather pattern
high_temp = moscow.loc[pd.datetime(2014,1,1) : pd.datetime(2014,12,31)]
```

```python
#were there any temperatures above 24 from Jan to December?lets see
#5 days were above 24 degrees
high_temp[high_temp['Mean TemperatureC'] >= 24]
```
